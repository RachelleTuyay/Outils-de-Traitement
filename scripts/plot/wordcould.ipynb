```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import glob, os
```

```python
cleaned_docs = "../../data/clean"

# Charger tous les fichiers .txt dans une liste
file_paths = glob.glob(os.path.join(cleaned_docs, "*.txt"))
documents = []
for path in file_paths:
    with open(path, encoding='utf-8') as f:
        documents.append(f.read())

text = " ".join(documents)
```

```python
#Génére le wc
wc = WordCloud(width=800, height=400, background_color='white').generate(text)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title("Nuage de mots du corpus", fontsize=16)

plt.savefig("../../figures/wordcloud.png", bbox_inches='tight')
plt.show()
```
