```python
import os
import glob, re
import matplotlib.pyplot as plt
```

```python

# Dossier contenant les fichiers texte
corpus_clean = "../../data/clean"  # À modifier selon ton dossier
corpus_raw = "../../data/raw"

# Récupérer tous les fichiers .txt
file_paths = glob.glob(os.path.join(corpus_clean, "*.txt"))

# Calculer la longueur de chaque fichier (en nombre de caractères)
file_lengths = []
file_names = []

for path in file_paths:
    with open(path, encoding='utf-8') as f:
        text = f.read()
    file_lengths.append(len(text))
    
    # Extraire que le numéro du fichier
    base_name = os.path.basename(path)
    match = re.search(r'\d+', base_name)
    file_names.append(match.group() if match else base_name)
```

```python
# Tracer le graphique
plt.figure(figsize=(10, 6))
plt.barh(file_names, file_lengths, color='red')
plt.xlabel("Longueur du fichier (nombre de caractères)")
plt.title("Longueur des fichiers texte du corpus clean")
plt.tight_layout()

plt.savefig("../../figures/lenght_mean_clean.png", bbox_inches='tight')
plt.show()
```
